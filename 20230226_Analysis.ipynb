```python
import pandas as pd
import numpy as np

import plotly
import plotly.graph_objs as go 
import plotly.offline as offline
import plotly.express as px
```

```python
def searchdf(df, **criteria):
    for col, val in criteria.items():
        df=df[df[col]==val]
    return df
```

```python
df_bothunter = pd.read_csv('./02_cleaned/bothunter.csv').iloc[:,1:]
df_bothunter
```

```python
## Reclassify bots with threshold 0.7
## what is the current threshold
print(f"current threshold to classify as bot = {df_bothunter[df_bothunter['prediction']==1]['probability'].min()}")
```

```python
df_bothunter.loc[df_bothunter['probability']>=0.7, 'prediction']=1
df_bothunter.loc[df_bothunter['probability']<0.7, 'prediction']=0
print(f"current threshold to classify as bot = {df_bothunter[df_bothunter['prediction']==1]['probability'].min()}")
df_bothunter
```
