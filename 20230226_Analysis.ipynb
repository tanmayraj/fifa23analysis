```python
import pandas as pd
import numpy as np

import plotly
import plotly.graph_objs as go 
import plotly.offline as offline
import plotly.express as px
```

```python
def searchdf(df, **criteria):
    for col, val in criteria.items():
        df=df[df[col]==val]
    return df
```

<h2> Read Data

```python
df_bothunter = pd.read_csv('./02_cleaned/bothunter.csv').iloc[:, 1:]
df_bothunter
```

```python
## Reclassify bots with threshold 0.7
## what is the current threshold
print(f"current threshold to classify as bot = {df_bothunter[df_bothunter['prediction']==1]['probability'].min()}")
```

```python
df_bothunter.loc[df_bothunter['probability']>=0.7, 'prediction']=1
df_bothunter.loc[df_bothunter['probability']<0.7, 'prediction']=0
print(f"current threshold to classify as bot = {df_bothunter[df_bothunter['prediction']==1]['probability'].min()}")
df_bothunter
```

```python
df_location = pd.read_csv('./02_cleaned/location.csv').iloc[:, 1:]
df_location
```

```python
df_emotion = pd.read_csv('./02_cleaned/emotion.csv', low_memory=False).iloc[:, 1:]
df_emotion
```

```python
df_morals = pd.read_csv('./02_cleaned/morals.csv').iloc[:, 1:]
df_morals
```

<h2>Data Cleaning</h2>

```python
## Additional cleaning - Removing userids with string
display(df_emotion[df_emotion['userid'].str.isnumeric()==False])
df_emotion = df_emotion[df_emotion['userid'].str.isnumeric()==True]
df_emotion['userid'] = df_emotion['userid'].astype(int)
df_emotion
```

```python
df_location
```

```python
df_location['region']
```

## Bothunter + Location

```python
df_bothunter_location = df_location.merge(df_bothunter, on='status_id')
df_bothunter_location
```

```python
## % of bots vs humans distributed across regions
df_bothunter_freq = pd.DataFrame(df_bothunter_location.groupby(
    ['location', 'region', 'lat', 'lon', 'prediction']).size()).reset_index().rename(columns={0:'count'})
df_bothunter_freq
df_bothunter_freq.loc[:,'percentage']=(df_bothunter_freq['count']*100/df_bothunter_freq.groupby(['location', 'region', 'lat', 'lon'])['count'].transform('sum')).round(2)
df_bothunter_freq
```

```python
df_bothunter_freq.to_csv('./03_results/bothunter_location.csv')
```

#### Approach : Lets take aggregate emotions of each userid

```python
df_emotion_mean = df_emotion.groupby('userid').mean().reset_index()
df_emotion_mean
```

```python
searchdf(df_emotion, userid=1984)
```

#### Merge (bothunter + Location is the base table)

```python
display(df_bothunter_location['user_id'].map(df_emotion_mean['userid'].value_counts()))
```

```python
df_bothunter_location_emotion = df_bothunter_location.merge(df_emotion_mean, left_on='user_id', right_on='userid')
df_bothunter_location_emotion
```

```python
df_ble_freq = \
pd.DataFrame(df_bothunter_location_emotion.groupby(
    ['location', 'region', 'lat', 'lon', 'prediction']).agg({'anger':'mean', 
                                                             'disgust':'mean',
                                                             'fear':'mean',
                                                             'joy':'mean',
                                                             'neutral':'mean',
                                                             'sadness':'mean',
                                                             'surprise':'mean'}).reset_index())
df_ble_freq
```

```python
df_ble_freq.to_csv('./03_results/bothunter_location_emotions.csv')
```
